```python
import os
from modify_structure_spglib import check_symmetry

# Define the base directory
directory = "/mnt/lustre-grete/usr/u12658/MOF-5/CP2K/strain/PBE_new_xyz/2cluster"

# Ensure the directory exists before proceeding
if not os.path.exists(directory):
    raise FileNotFoundError(f"Error: The specified directory does not exist: {directory}")

# List only valid folders starting with "comp_" or "exp_"
folders = [f for f in os.listdir(directory) if (f.startswith('comp_') or f.startswith('exp_'))]

# Process each folder
for fold in folders:
    print(f"Processing folder: {fold}")
    folder_path = os.path.join(directory, fold)
    
    if not os.path.isdir(folder_path):  # Ensure it's a valid directory
        continue

    xyz_file_path = os.path.join(folder_path, "new.xyz")

    if os.path.exists(xyz_file_path):  # Check if new.xyz exists
        try:
            check_symmetry([xyz_file_path])  # Pass the file as a list
            print(f"Processed: {xyz_file_path}")  # Confirmation message
        except Exception as e:
            print(f"Error processing {xyz_file_path}: {e}")
    else:
        print(f"Warning: {xyz_file_path} not found.")
```
